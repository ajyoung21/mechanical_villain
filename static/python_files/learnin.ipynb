```python
import pandas as pd
import xgboost
```

```python
df_price = pd.read_csv("../data/with_price.csv")
df_noprice = pd.read_csv("../data/without_price.csv")
```

```python
df_train = df_price.drop(["Island_Name", "Country", "nearest_city"], axis=1)
df_train
```

```python
df_test = df_noprice.drop(["price", "Island_Name", "Country", "nearest_city"], axis=1)
df_test
```

```python
x_train = df_train.drop(["price"], axis=1)
y_train = df_train["price"]
```

```python
classifier = xgboost.XGBRegressor()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(df_test)
```

```python
df_noprice["price"] = y_pred
```

```python
df_noprice
```

```python
df_noprice["predicted"] = "yes"
df_price["predicted"] = "no"
df_final = df_price.append(df_noprice)
df_final.to_csv("../data/island_final_price.csv")
```
