```python
import pandas as pd
df = pd.read_csv("../data/island_info.csv")
df
```

```python
new_dict = {}
attributes = []

for column in df:
    attributes = []
    column_list = df[column].tolist()
    for item in column_list:
        if item == "yes":
            attribute = 1
        elif item == "no":
            attribute = 0
        else:
            attribute = item
        attributes.append(attribute)
    
    new_dict[column] = attributes

new_df = pd.DataFrame(new_dict)
df = new_df
```

```python
df2 = df.dropna(subset=["price"])
df2.to_csv("../data/with_price.csv", index=False)
```

```python
price_list = df["price"].tolist()
counter = 0
index_list = []

for item in price_list:
    if item > 1:
        index_list.append(counter)
    else:
        pass
    
    counter = counter+1

index_list.reverse()

nan_dict = {}

for column in df:
    column_list = df[column].tolist()
    for item in index_list:
        column_list.pop(item)
    nan_dict[column] = column_list

df3 = pd.DataFrame(nan_dict)
df3.to_csv("../data/without_price.csv", index=False)
    
```
