```python
import pandas as pd
```

```python
df = pd.read_csv("../data/island_info.csv")
df
```

```python
price_list = df["price"].tolist()
price_list
prices = []
for item in price_list:
    check = item[0:3]
    if check == "USD":
        price = item[3:]
    else:
        price = "NA"
    prices.append(price)

df2 = df.drop(columns=["price"])
df2["price"]=prices
df2.to_csv("../data/island_info.csv")
```
