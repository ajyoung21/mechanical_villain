```python
import pandas as pd
import scipy.stats as st
```

```python
price_df = pd.read_csv("../data/with_price.csv")
corruption_df = pd.read_csv("../data/merged_countries.csv")
corruption_df
```

```python
distance_mean = price_df['city_distance'].mean()
distance_mean
```

```python
distance_std = price_df['city_distance'].std()
distance_std
```

```python
price_mean = price_df['price'].mean()
price_mean
```

```python
price_std = price_df['price'].std()
price_std
```

```python
price_df['Z_Price'] = (price_df['price'] - price_mean)/price_std
price_df
```

```python
refined = price_df['Acreage'].replace(0,1)
price_df['Acreage'] = refined
price_df
```

```python
price_df['$/Acre'] = price_df['price']/price_df['Acreage']
price_df
```

```python
value_mean = price_df['$/Acre'].mean()
value_std = price_df['$/Acre'].std()
price_df['Z_Value'] = (price_df['$/Acre'] - value_mean)/value_std
price_df
```

```python
price_df['P_Value'] = st.norm.cdf(price_df['Z_Value'])
price_df.head()
```

```python
price_df['Z_Distance'] = (price_df['city_distance'] - distance_mean)/distance_std
price_df.head()
```

```python
price_df['P_Distance'] = st.norm.cdf(price_df['Z_Distance'])
price_df.head()
```

```python
price_df = price_df.merge(corruption_df, on='Country')
price_df.head()
```

```python
homi_mean = price_df['Homicide Rate'].mean()
homi_std = price_df['Homicide Rate'].std()

cpi_mean = price_df['CPI Score 2018'].mean()
cpi_std = price_df['CPI Score 2018'].std()
```

```python
price_df['Z_Homicide'] = (price_df['Homicide Rate'] - homi_mean)/homi_std
price_df['P_Homicide'] = st.norm.cdf(price_df['Z_Homicide'])
price_df.head()
```

```python
price_df['Z_CPI'] = (price_df['Homicide Rate'] - cpi_mean)/cpi_std
price_df['P_CPI'] = st.norm.cdf(price_df['Z_CPI'])
price_df['Income_Potential'] = price_df['Income Potential']
```

```python
price_df.to_csv("../data/standardized_info.csv", index=False)
```
