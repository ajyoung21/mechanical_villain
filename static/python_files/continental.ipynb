```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("../data/island_final_price.csv")
df.head()
```

```python
country_list = df.Country.unique()
country_list
```

```python
asia = ['Thailand', 'United Arab Emirates', 'Maldives', 'Philippines', 'Japan', 'Indonesia']
europe = ['France', 'Greece', 'United Kingdom', 'Ireland', 'Italy', 'Finland', 'Croatia', 'Estonia', 'Poland', 'Iceland']
africa = ['Madagascar', 'Uganda', 'Seychelles']
n_america = ['United States of America', 'Bahamas', 'Panama', 'Belize', 'Grenada', 'Mexico', 'Canada', 'Curacao', 'Antigua',
            'Honduras', 'Nicaragua', 'Costa Rica', 'Turks and Caicos', 'SVG']
s_america = ["Chile", 'Brazil', 'Argentina', 'Ecuador', 'Falkland Islands']
oceania = ['Fiji', 'Solomon Islands', 'New Zealand', 'Australia', 'Vanuatu', 'Tonga', 'Papua New Guinea', 'Palau']
```

```python
continents = dict({'Asia':asia, 'Europe':europe, 'Africa':africa, 'North America': n_america, 'South America': s_america, 'Oceania': oceania})
```

```python
rev_continents = {}
for k,v in continents.items():
    for x in v:
        rev_continents.setdefault(x,[]).append(k)
```

```python
print(rev_continents)
```

```python
df['Continent'] = df['Country'].map(rev_continents) 
df.head()
```

```python
df['Continent'] = df['Continent'].str[0]
df.head()
```

```python
df2 = df.loc[:, ['Continent', 'price']]
df2.head()
```

```python
df2 = df2.sort_values('Continent')
df2.head()
```

```python
africa2 = df.loc[df.Continent=='Africa']
asia2 = df.loc[df.Continent=='Asia']
europe2 = df.loc[df.Continent=='Europe']
n_america2 = df.loc[df.Continent=='North America']
s_america2 = df.loc[df.Continent=='South America']
oceania2 = df.loc[df.Continent=='Oceania']
```

```python
africa2 = africa2.loc[:, ['Continent', 'price']]
asia2 = asia2.loc[:, ['Continent', 'price']]
europe2 = europe2.loc[:, ['Continent', 'price']]
n_america2 = n_america2.loc[:, ['Continent', 'price']]
s_america2 = s_america2.loc[:, ['Continent', 'price']]
oceania2 = oceania2.loc[:, ['Continent', 'price']]
```

```python
af_price = africa2['price'].mean()
as_price = asia2['price'].mean()
eu_price = europe2['price'].mean()
na_price = n_america2['price'].mean()
sa_price = s_america2['price'].mean()
oc_price = oceania2['price'].mean()
print(af_price)
print(oc_price)
print(sa_price)
print(na_price)
print(eu_price)
print(as_price)
```
